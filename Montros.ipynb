```python
import requests
import pandas as pd
```

```python
base_url = 'https://www.dnd5eapi.co'
url = f'{base_url}/api/monsters'
response = requests.get(url)
monstros_json = response.json()
monstros_json['results'][0]['name']
```

```python
monstros = pd.DataFrame()
temp = []
urls = []
for monstro in monstros_json['results']:
    temp.append(monstro['name'])
    urls.append(monstro['url'])

monstros['name'] = temp
monstros

monstros.to_csv('nomes_monstros')
```

```python
df = pd.DataFrame(columns=[
    'nome',
    'tamanho',
    'tipo',
    'alinhamento',
    'armadura',
    'vida',
    'velocidade',
    'forca',
    'destreza',
    'constituicao',
    'inteligencia',
    'intuicao',
    'carisma',
    'vulnerabilidade',
    'resistencias',
    'imunidades',
    'idioma',
    'habilidades especiais',
    'habilidades',
    'xp',
    'locais'
])

df
```

```python
nome = []
tamanho = []
tipo = []
alinhamento = []
armadura = []
vida = []
velocidade = []
forca = []
destreza = []
constituicao = []
inteligencia = []
intuicao = []
carisma = []
vulnerabilidade = []
resistencias = []
imunidades = []
idioma = []
habilidades_especiais = []
habilidades = []
xp = []
locais = []
```

```python
for link in monstros['url']:
    url = base_url+link
    response = requests.get(url)
    pagina_monstro = response.json()
    nome.append(pagina_monstro['name'])
    tamanho.append(pagina_monstro['size'])
    tipo.append(pagina_monstro['type'])
    alinhamento.append(pagina_monstro['alignment'])
    armadura.append(pagina_monstro['armor_class'][0]['value'])
    vida.append(pagina_monstro['hit_points'])
    velocidade.append(pagina_monstro['speed'])
    forca.append(pagina_monstro['strength'])
    destreza.append(pagina_monstro['dexterity'])
    constituicao.append(pagina_monstro['constitution'])
    inteligencia.append(pagina_monstro['intelligence'])
    intuicao.append(pagina_monstro['wisdom'])
    carisma.append(pagina_monstro['charisma'])
    vulnerabilidade.append(pagina_monstro['damage_vulnerabilities'])
    resistencias.append(pagina_monstro['damage_resistances'])
    idioma.append(pagina_monstro['languages'])
    imunidades.append(pagina_monstro['damage_immunities'])
    habilidades_especiais.append(pagina_monstro['special_abilities'])
    habilidades.append(pagina_monstro['ac'])
    xp.append(pagina_monstro['xp'])



    
```

```python
df['nome'] = nome
df['tamanho'] = tamanho
df['tipo'] = tipo
df['alinhamento'] = alinhamento
df['armadura'] = armadura
df['vida'] = vida
df['velocidade'] = velocidade
df['forca'] = forca
df['destreza'] = destreza
df['constituicao'] = constituicao
df['inteligencia'] = inteligencia
df['intuicao'] = intuicao
df['carisma'] = carisma
df['vulnerabilidade'] = vulnerabilidade
df['resistencias'] = resistencias
df['imunidades'] = imunidades
df['idioma'] = idioma
df['habilidades especiais'] = habilidades_especiais
df['habilidades'] = habilidades
df['xp'] = xp
```
